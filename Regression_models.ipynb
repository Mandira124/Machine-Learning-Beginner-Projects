```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import PolynomialFeatures

# Dataset
X = np.array([1, 2, 3, 4, 5, 6, 7, 8]).reshape(-1, 1)
y = np.array([30, 38, 50, 58, 62, 68, 80, 95])

# Linear Regression
linear_model = LinearRegression() # 1 extra hour of study increases marks by a constant amount
linear_model.fit(X, y)

# Polynomial Regression
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)      # model is curved, useful when the relationship between X and Y is dynamic
poly_model = LinearRegression()
poly_model.fit(X_poly, y)

# Ridge Regression
ridge_model = Ridge(alpha=1)        # regularized version of linear regression, avoids weird data i.e avoids extreme slopes
ridge_model.fit(X, y)

# Lasso Regression
lasso_model = Lasso(alpha=0.1)      # removes features that are not that important , depend on alpha....how aggressive the lasso regression is
lasso_model.fit(X, y)

# Range for smooth plotting
X_range = np.linspace(1, 8, 100).reshape(-1, 1)

# Plotting
plt.scatter(X, y)
plt.plot(X_range, linear_model.predict(X_range), label="Linear Regression")
plt.plot(X_range, poly_model.predict(poly.transform(X_range)), label="Polynomial Regression")
plt.plot(X_range, ridge_model.predict(X_range), label="Ridge Regression")
plt.plot(X_range, lasso_model.predict(X_range), label="Lasso Regression")

plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.title("Comparison of Regression Models")
plt.legend()
plt.show()

# Print coefficients
print("Linear coefficients:", linear_model.coef_)
print("Ridge coefficients:", ridge_model.coef_)
print("Lasso coefficients:", lasso_model.coef_)
```
